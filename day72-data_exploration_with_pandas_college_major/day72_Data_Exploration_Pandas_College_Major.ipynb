```python
clean_df = df.dropna()
clean_df.tail()
```

Highest starting median salary

```python
idx_max_starting_med_slry = clean_df['Starting Median Salary'].idxmax()
clean_df.loc[idx_max_starting_med_slry]
```

Highest mid-career median salary

```python
idx_max_mid_career_med_slry = clean_df['Mid-Career Median Salary'].idxmax()
clean_df.loc[idx_max_mid_career_med_slry]
```

Lowest starting median salary

```python
idx_min_starting_med_slry = clean_df['Starting Median Salary'].idxmin()
clean_df.loc[idx_min_starting_med_slry]
```

Lowest mid-career median salary

```python
idx_min_mid_career_med_slry = clean_df['Mid-Career Median Salary'].idxmin()
clean_df.loc[idx_min_mid_career_med_slry]
```

Lowest risk majors

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary'] - clean_df['Mid-Career 10th Percentile Salary']
# clean_df.insert(1, "Spread", spread_col)
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

Highest risk majors

```python
high_risk = clean_df.sort_values('Spread', ascending=False)
high_risk[['Undergraduate Major', 'Spread']].head()
```

Top 5 degrees with the highest values in the 90th percentile

```python
top_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
top_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```
